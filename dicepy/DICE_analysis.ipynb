## Utility Functions

```python
def utility(c, alpha):
    u = (np.power(c,(1-alpha))-1)/(1-alpha)-1
    return u
```

```python
c=np.linspace(10, 1000, 100)
```

```python
u = utility(c, 0.5)
```

```python
plt.plot(c, utility(c, 0.75), label="0.5");
plt.plot(c, utility(c, 0.9), label="0.9");
plt.plot(c, utility(c, 0.99), label="0.99");
plt.plot(c, utility(c, 1.1), label="1.1");
plt.xlabel("Consumption")
plt.ylabel("Utility")
plt.legend()
plt.savefig("./figures/UtilityFunctions.jpeg")
```
